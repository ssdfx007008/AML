```python
%matplotlib inline

import numpy as np
from scipy.spatial import distance
from matplotlib import pyplot as plt
from sklearn import manifold
from sklearn.decomposition import PCA
import _pickle as cPickle
from collections import defaultdict
```

```python
def unpickle(file):
    fo = open(file, 'rb')
    dict = cPickle.load(fo,  encoding='latin1')
    fo.close()
    return dict
```

```python
batch_data = []
batch_label = []
for i in range(1,6):
    print("reading file " + str(i) + "...")
    batch_data.append(unpickle("cifar-10-batches-py/data_batch_" + str(i))["data"])
    batch_label.append(unpickle("cifar-10-batches-py/data_batch_" + str(i))["labels"])
    
```

```python
category = defaultdict(list)
```

```python
for i in range(5):
    for j in range(len(batch_data[i])):
        category[batch_label[i][j]].append(batch_data[i][j])
```

```python
mean_images = []

for i in range(10):
    mean_image= np.zeros(len(category[i][0]),dtype=int)
    for j in range(len(category[i])):
        mean_image += np.asarray(category[i][j],dtype=int)
    mean_image = np.divide(mean_image, len(category[i]))
    mean_image = np.array(mean_image, 'uint8')
    mean_images.append(mean_image)
```

```python

for i in range(len(mean_images)):
    plt.figure()
    plt.imshow(np.reshape(mean_images[i], (32,32,3), order='F'))

    plt.title("Mean image for category {}".format(i))
```
