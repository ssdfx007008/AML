```python
%matplotlib inline

import numpy as np
from scipy.spatial import distance
from matplotlib import pyplot as plt
from sklearn import manifold
from sklearn.decomposition import PCA
import _pickle as cPickle
from collections import defaultdict
```

```python
def unpickle(file):
    fo = open(file, 'rb')
    dict = cPickle.load(fo,  encoding='latin1')
    fo.close()
    return dict
```

```python
batch_data = []
batch_label = []
for i in range(1,6):
    print("reading file " + str(i) + "...")
    batch_data.append(unpickle("cifar-10-batches-py/data_batch_" + str(i))["data"])
    batch_label.append(unpickle("cifar-10-batches-py/data_batch_" + str(i))["labels"])
    
```

```python
category = defaultdict(list)
```

```python
for i in range(5):
    for j in range(len(batch_data[i])):
        category[batch_label[i][j]].append(batch_data[i][j])
```

```python
mean_images = []
for i in range(10):
    mean_image= [0 for _ in range(len(category[i][0]))]
    for j in range(len(category[i])):
        mean_image = [x+y for x, y in zip(mean_image, category[i][j])]
    mean_image = [int(x/len(category[i])) for x in mean_image]
    im = np.reshape(mean_image, (32,32,3), order='F')
    mean_images.append(im)
```

```python
for i in range(len(mean_images)):
    plt.figure()
    plt.imshow(mean_images[i])
        print("Mean image for category {}".format(i))
```
